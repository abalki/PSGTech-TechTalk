# **Descriptive Analytics - Linear Regression example**

**Imports**

```
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

**Calculate coefficients**

```
def estimate_coefficients(x, y): 
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    variance = np.var(x, ddof=1)
    covariance = np.cov(x, y, bias=True)[0][1]
    b1 = covariance / variance
    b0 = y_mean - b1 * x_mean
    return(b0, b1)
```

**Plot graph**

```
# x,y are the location of points on graph

def plot_regression_line(x, y, b): 
    # plotting the points as per dataset on a graph
    plt.scatter(x, y, color = "m",marker = "o", s = 30) 

    # predicted response vector 
    y_pred = b[0] + b[1]*x 
  
    # plotting the regression line
    plt.plot(x, y_pred, color = "g")
  
    # putting labels for x and y axis
    plt.xlabel('Price (Rs.)') 
    plt.ylabel('Demand (Units)') 
  
    # function to show plotted graph
    plt.show()
```

**Plot seaborn**

```
def plot_seaborn(x, y): 
    sns.set(color_codes=True)
    ax = sns.regplot(x=x, y=y, color="g")
```

**Perform actual regression**

```
# Datasets which we create 
x = np.array([4.0, 3.5, 5.0, 6.0, 6.5, 7.0, 2.0, 4.0, 5.5, 3.0, 3.5, 2.0, 2.0, 
              3.0, 3.0, 1.5, 3.0, 4.8, 5.0, 4.0, 4.5, 4.0, 7.5, 4.0, 6.5, 4.0, 
              7.0, 5.5, 7.0, 3.5, 7.0, 2.0]) 
y = np.array([7, 7, 5, 4, 4, 4, 8, 6, 5, 7, 7, 8, 8, 8, 7, 9, 8, 5, 5, 7, 7, 8,
              3, 7, 4, 7, 3, 5, 5, 7, 5, 8])
  
# estimating coefficients 
b = estimate_coefficients(x, y) 
print("Estimated coefficients:\nb_0 = {} \nb_1 = {}".format(b[0], b[1])) 
  
# plotting regression line 
plot_regression_line(x, y, b)

# Plot using seaborn
plot_seaborn(x, y)
```
